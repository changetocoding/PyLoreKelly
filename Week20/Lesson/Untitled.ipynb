```python
import pandas as pd
download_url = (
    "https://raw.githubusercontent.com/fivethirtyeight/"
    "data/master/college-majors/recent-grads.csv"
)
df = pd.read_csv(download_url)
type(df)
```

```python
df.head()
```

```python
%matplotlib inline

df.plot(x="Rank", y=["P25th", "Median", "P75th"])
```

```python
median_column = df["Median"]
type(median_column)  # pandas.core.series.Series
median_column.plot(kind="hist")
```

```python
top_5 = df.sort_values(by="Median", ascending=False).head()
top_5
```

```python

top_5.plot(x="Major", y="Median", kind="bar", rot=90, fontsize=12)
```

```python
medians = df[df["Median"] > 60000]

medians.plot(x="Major", y=["P25th", "Median", "P75th"], kind="bar")
```

```python
df.plot(x="Median", y="P75th", kind="scatter")
```

```python
df.corr()
```

```python
cat_totals = df.groupby("Major_category")["Median"].mean().sort_values()
cat_totals
```

```python

cat_totals.plot(kind="barh", fontsize=12)
```

```python
c_totals = df.groupby("Major_category")["Total"].sum().sort_values()
small_cat_totals = c_totals[c_totals < 200_000]
big_cat_totals = c_totals[c_totals > 200_000]

small_cat_totals
```

```python
small_sums = pd.Series([small_cat_totals.sum()], index=["Other"])
big_cat_totals = big_cat_totals.append(small_sums)
```

```python
big_cat_totals.plot(kind="pie", label="")
```
